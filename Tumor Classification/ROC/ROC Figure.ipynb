```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load fpr and tpr arrays for each model
fpr_model1 = np.load('./fpr-VGG.npy')
tpr_model1 = np.load('./tpr-VGG.npy')
fpr_model2 = np.load('./fpr-Res.npy')
tpr_model2 = np.load('./tpr-Res.npy')
fpr_model3 = np.load('./fpr-X.npy')
tpr_model3 = np.load('./tpr-X.npy')
```

```python
# Plot ROC curves for each model
plt.figure()
plt.plot(fpr_model1, tpr_model1, color='orange', lw=2,linestyle='-', label="VGG2D")
plt.plot(fpr_model2, tpr_model2, color='blue', lw=2, linestyle='--' ,label="ResNet50V2")
plt.plot(fpr_model3, tpr_model3, color='green', lw=2, linestyle=':', label="Xception")
plt.plot([0, 1], [0, 1], color='navy', lw=1, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc="lower right")
plt.show()
```

```python
# Plot ROC curves for each model with more detail
plt.figure()
plt.plot(fpr_model1, tpr_model1, color='darkorange', lw=2,linestyle='-', label="VGG2D")
plt.plot(fpr_model2, tpr_model2, color='blue', lw=2, linestyle='--' ,label="ResNet50V2")
plt.plot(fpr_model3, tpr_model3, color='green', lw=2, linestyle=':', label="Xception")
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 0.5])  # Zoom into the region where the curve is changing
plt.ylim([0.8, 1.0])  # Zoom into the region where the curve is changing
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves (Zoomed)')
plt.legend(loc="lower right")
plt.show()
```
